```python
# Show unique values for column 'age'

dataframe['age'].unique()

# convert 'age' to int
dataframe['age'] = dataframe['age'].astype(int)

# Remove age from features_categorical
features_categorical = features_categorical.drop(columns=['age'], axis=1)

# Add 'age' to features_numerical
features_numerical['age'] = dataframe['age'] 
```

```python
# Replace `?` with `undefined` in dataframe
dataframe = dataframe.replace(' ?', 'undefined')

# Show how many columns have `?` in the dataframe
for col in features_categorical.columns.to_list():
    print(f'{col}: {dataframe[col].str.contains("\?").sum()}')
```

```python
# Replace `undefined` with the most frequent value in each column or mode

for col in features_categorical.columns.to_list():
    dataframe[col] = dataframe[col].replace('undefined', dataframe[col].mode()[0])

# Show unique values for each categorical column
for col in features_categorical.columns.to_list():
    print(f'{col}: {dataframe[col].unique()}')
```
