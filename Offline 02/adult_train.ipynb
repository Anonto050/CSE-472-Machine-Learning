# Understanding dataset

```python
import pandas as pd

columns = columns = [
    'age', 'workclass', 'fnlwgt', 'education', 'education-num',
    'marital-status', 'occupation', 'relationship', 'race', 'sex',
    'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'income'
]

dataframe = pd.read_csv('adult.data', names=columns)
dataframe.head(10)
```

```python
# Check for empty string ("") in any of the columns
empty_string_mask = (dataframe == "")

# Print the columns and rows where empty strings are found
if empty_string_mask.any().any():
    print("Empty string found in the dataset:")
    print(empty_string_mask.sum())  # Print number of empty strings in each column
    print("Rows with empty strings:")
    print(dataframe[empty_string_mask.any(axis=1)])  # Print rows containing empty strings
else:
    print("No empty strings found in the dataset.")
```

# Numerical vs Categorical

```python
features_numerical = dataframe.select_dtypes(include=['int64', 'float64'])
features_numerical.columns
```

```python
features_categorical = dataframe.select_dtypes(include=['object'])
features_categorical.columns
```

```python
# Histograms for numerical features

import matplotlib.pyplot as plt

features_numerical.hist(figsize=(10, 10))
plt.show()
```

```python
import seaborn as sns

fig, axes = plt.subplots(9, 1, figsize=(20, 15))
for i, col in enumerate(features_categorical):
    sns.countplot(x=col, data=features_categorical, ax=axes[i])
plt.subplots_adjust(hspace=0.5)
```

```python
# Show how many categorical columns have a value of `?`
for col in features_categorical.columns.to_list():
    print(f'{col}: {features_categorical[col].str.contains('\?').sum()}')
```

```python
# Replace `?` with `undefined`

dataframe = dataframe.replace(' ?', 'undefined')

# Show how many categorical columns have a value of `?`
for col in features_categorical.columns.to_list():
    print(f'{col}: {dataframe[col].str.contains('\?').sum()}')
```

```python
# Replace `undefined` with the most frequent value in each column or mode

for col in features_categorical.columns.to_list():
    dataframe[col] = dataframe[col].replace('undefined', dataframe[col].mode()[0])

# Show unique values for each categorical column
for col in features_categorical.columns.to_list():
    print(f'{col}: {dataframe[col].unique()}')
    
```
