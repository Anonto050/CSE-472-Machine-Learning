# Understanding dataset

```python
# Load the dataset

dataframe = pd.read_csv('creditcard.csv')
dataframe.head(10)
```

```python
dataframe.shape
# (284807, 31)

dataframe.columns
dataframe.dtypes
dataframe.describe()
dataframe.dtypes.value_counts()
```
