# Data Cleaning

```python
if employee_dataframe.isnull().values.any():
    employee_dataframe.fillna(employee_dataframe.mean(numeric_only=True), inplace=True)

employee_dataframe.dropna(subset=['Attrition'], inplace=True)
```

# Conversion of features into numeric values

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
Labels = employee_dataframe['Attrition']
Labels = encoder.fit_transform(Labels)

Labels
```

```python

def one_hot_encode(Features):
    categorical_columns = Features.select_dtypes(exclude=['int64', 'float64']).columns
    Features[categorical_columns] = Features[categorical_columns].astype('category')

    Features_encoded = pd.get_dummies(Features, columns=categorical_columns)
    
    boolean_columns = Features_encoded.select_dtypes(include=['bool']).columns
    Features_encoded[boolean_columns] = Features_encoded[boolean_columns].astype('int')

    return Features_encoded

Features_encoded = one_hot_encode(Features)
Features_encoded

    
```

```python
Features_encoded.dtypes
```

# Scaling of the Features

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

def scale_features(Features, scaling_method='minmax'):
    """
    This function scales the numerical features of the provided Features based on the chosen scaling method.
    
    Parameters:
    - Features: The input DataFrame containing the features to be scaled (numerical features).
    - scaling_method: A string that defines the scaling method ('minmax' or 'standard'). Default is 'minmax'.
    
    Returns:
    - Scaled numerical features as a DataFrame.
    """
    # Select numerical columns
    numerical_cols = Features.select_dtypes(include=['int64', 'float64']).columns

    # Initialize the scaler based on the chosen method
    if scaling_method == 'minmax':
        scaler = MinMaxScaler()
    elif scaling_method == 'standard':
        scaler = StandardScaler()
    else:
        raise ValueError('Scaling method not recognized. Use "minmax" or "standard".')

    # Apply the scaler to the numerical columns
    scaled_features = scaler.fit_transform(Features[numerical_cols])

    # Convert the scaled features back to a DataFrame and return it
    scaled_features_df = pd.DataFrame(scaled_features, columns=numerical_cols, index=Features.index)

    Features[numerical_cols] = scaled_features_df

    return Features

scaled_features_minmax = scale_features(Features, scaling_method='minmax')
scaled_features_minmax
    
    
```

```python
scaled_features_standard = scale_features(Features, scaling_method='standard')
scaled_features_standard
```

```python
scaled_features_minmax = scale_features(Features, scaling_method='minmax')
scaled_features_minmax = one_hot_encode(scaled_features_minmax)
```

```python
scaled_features_standard = scale_features(Features, scaling_method='standard')
scaled_features_standard = one_hot_encode(scaled_features_standard)
scaled_features_standard
```

# Correlation Analysis

```python
Labels_df = pd.DataFrame(Labels, columns=['Attrition'])
```

```python
# correlation analysis with features of minmax scaled data

Labels_series = Labels_df['Attrition']
correlations_minmax = scaled_features_minmax.corrwith(Labels_series)
correlations_minmax
```

```python
correlations_standard = scaled_features_standard.corrwith(Labels_series)
correlations_standard
```

```python
sorted_correlations_minmax = correlations_minmax.abs().sort_values(ascending=False)
sorted_correlations_minmax = sorted_correlations_minmax.head(20)
sorted_correlations_minmax
```

```python
sorted_correlations_standard = correlations_standard.abs().sort_values(ascending=False)
sorted_correlations_standard = sorted_correlations_standard.head(20)
sorted_correlations_standard
```

```python
matrix = scaled_features_minmax.corr()
matrix
```

```python
top_features = sorted_correlations_standard.index
matrix = scaled_features_standard[top_features].corr()
matrix
```

```python
import seaborn as sns

sns.heatmap(matrix, cmap='coolwarm', center=0, annot=False)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def scatter_plots(scaled_features_minmax, Labels_df, sorted_correlations_minmax):
    top_features = sorted_correlations_minmax.index

    # target_values = Labels_df.squeeze()
    # classes = target_values.unique()

    class_0 = scaled_features_minmax.loc[Labels_df['Attrition'] == 0]
    class_1 = scaled_features_minmax.loc[Labels_df['Attrition'] == 1]


    for feature in top_features:
        plt.figure(figsize=(10, 4))

        plt.plot(class_0[feature], np.zeros_like(class_0[feature]), 'o', label='Class 0')
        plt.plot(class_1[feature], np.zeros_like(class_1[feature])+1, 'o', label='Class 1')

        plt.legend()
        plt.xlabel(feature)
        plt.title(f'{feature} vs. Attrition')
        plt.show()

scatter_plots(scaled_features_minmax, Labels_df, sorted_correlations_minmax)
        
```

```python
# create a simple classifier
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(scaled_features_minmax[top_features], Labels, test_size=0.2, random_state=42)

model = LogisticRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)

print(f'Accuracy of logistic regression classifier: {accuracy:.2f}')
```
